```python
import pandas as pd
import numpy as np
```

```python
merge_keys = pd.read_excel("Standar Munis.xlsx")
```

```python
merge_keys
```

```python
macro_df = pd.read_excel("Base variables macro.xlsx")
```

```python
macro_df = macro_df.merge(merge_keys[["Estandár", "CGR"]].rename(columns={"CGR": "municipality"}), on="municipality", how="left")
macro_df = macro_df.drop(columns="municipality").rename(columns={"Estandár": "municipality"})
```

```python
cgr1_df = pd.read_csv("cgr_clean (Agregado).csv")
cgr2_df = pd.read_csv("cgr_clean (Agregado Medio).csv")
cgr3_df = pd.read_csv("cgr_clean (Desagregado).csv")
```

```python
cgr_df = cgr1_df.merge(cgr2_df, on=["year", "municipality"], how="left")
cgr_df = cgr_df.merge(cgr3_df, on=["year", "municipality"], how="left")
cgr_df = cgr_df.fillna(0)
```

```python
cgr_df = cgr_df.merge(merge_keys[["Estandár", "CGR"]].rename(columns={"CGR": "municipality"}), on="municipality", how="left")
cgr_df = cgr_df.drop(columns="municipality").rename(columns={"Estandár": "municipality"})
```

```python
mayors_df = pd.read_excel("elected_mayors_age_gender.xlsx", usecols=['municipality', 'year', 'type', 'Age', 'Gender'])
```

```python
mayors_df = mayors_df.merge(merge_keys[["Estandár", "elected_mayors"]].rename(columns={"elected_mayors": "municipality"}), on="municipality", how="left")
```

```python
mayors_df = mayors_df.drop(columns="municipality").rename(columns={"Estandár": "municipality"})
```

```python
mayors_df["Age"] = np.floor(mayors_df["Age"])
```

```python
elections_df = pd.read_excel("panel_abs_mar.xlsx").dropna()
```

```python
elections_df = elections_df.merge(merge_keys[["Estandár", "TSE"]].rename(columns={"TSE": "municipality"}), on="municipality", how="left")
elections_df = elections_df.drop(columns="municipality").rename(columns={"Estandár": "municipality"})
```

```python
population_df = pd.read_stata("population.dta")
```

```python
population_df["pop_share014"] = 100 * population_df["pop_share014"]
population_df["pop_share65plus"] = 100 * population_df["pop_share65plus"]
```

```python
population_df = population_df.merge(merge_keys[["Estandár", "CCSS"]].rename(columns={"CCSS": "municipality"}), on="municipality", how="left")
population_df = population_df.drop(columns="municipality").rename(columns={"Estandár": "municipality"})
```

```python
df = cgr_df.merge(mayors_df, on=["municipality", "year"], how='left')
df
```

```python
df = df.merge(elections_df, on=["municipality", "year"], how='left')
df
```

```python
df = df.merge(population_df, on=["municipality", "year"], how='left')
df
```

```python
df = df.merge(macro_df, on=["municipality", "year"], how='left')
df
```

```python
for col in df.columns: 
    print(col, df.loc[df[col] == 0.0, col].count())
```

```python
df.to_stata("base_regresiones.dta", write_index=False)
```
